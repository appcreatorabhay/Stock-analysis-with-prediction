```python
stockList=["RELIANCE.NS","TCS.NS","ADANIPORTS.NS","AAPL","GOOGL", "MSFT","AMZN","TSLA","ADBE","CSCO"]
symbols=" ".join(stockList)
```

```python
tickers=yf.Tickers(symbols)
```

```python
for ticker in stockList:
    print(ticker)
    for key,value in tickers.tickers[ticker].info.items():
        print(key,":",value)
    print('\n')
```

```python
all_data = []
# Loop through each ticker symbol
for ticker in stockList:
    print(f"Fetching data for: {ticker}")
    
    # Fetch the historical data for the last 5 years
    history = tickers.tickers[ticker].history(period="5y")
    
    # Optionally, you can include the ticker symbol as a column in the dataframe
    history['Ticker'] = ticker
    
    # Append the data to the list
    all_data.append(history)

# Concatenate all dataframes into a single one
final_data = pd.concat(all_data)

# Save the combined dataframe to a CSV file
final_data.to_csv("stock_data_5_years_final.csv")
```

```python
Close=history['Close']['RELIANCE.NS'].to_numpy()
Open=history['Open']['RELIANCE.NS'].to_numpy()
High=history['High']['RELIANCE.NS'].to_numpy()
Low=history['Low']['RELIANCE.NS'].to_numpy()
#for matplotlib.pyplot we provide data in the form of numpy array
```

```python
SMA=talib.SMA(Close,timeperiod=14) #can read more about SMA at https://www.investopedia.com/terms/s/sma.asp
SMA
```

```python
EMA=talib.EMA(Close,timeperiod=14) #can read more about EMA at https://www.investopedia.com/terms/e/ema.asp
EMA
```

```python
plt.figure(figsize=(30, 15))
plt.plot(history.index,Close,color="red",label="Close")
plt.plot(history.index,Open,color="green",label="Open")
plt.plot(history.index,SMA,color="yellow",label="SMA")
plt.legend()
plt.show()
#SMA EMA Crossover
plt.figure(figsize=(30, 10))
plt.plot(history.index,EMA,color="blue",label="EMA")
plt.plot(history.index,SMA,color="orange",label="SMA")
plt.legend()
plt.show()
```

```python
final_data = final_data.reset_index(drop=True)  # Ensure no MultiIndex is present

# Add technical indicators using pandas-ta
final_data['SMA_50'] = talib.SMA(final_data['Close'], timeperiod=50)  # Simple Moving Average with 50 periods
final_data['SMA_200'] = talib.SMA(final_data['Close'], timeperiod=200)  # Simple Moving Average with 200 periods
final_data['RSI'] = talib.RSI(final_data['Close'], timeperiod=14)  # Relative Strength Index with 14 periods
```

```python
# Visualization example
for ticker in stockList:
    plt.figure(figsize=(14, 7))  # Set figure size
    company_data = final_data[final_data['Ticker'] == ticker]  # Filter data for the specific ticker
    plt.plot(company_data.index, company_data['Close'], label=f'{ticker} Close Price')
    plt.plot(company_data.index, company_data['SMA_50'], label='50-Day SMA')
    plt.plot(company_data.index, company_data['SMA_200'], label='200-Day SMA')
    plt.plot(company_data.index, company_data['RSI'], label='RSI')

    # Add titles and labels
    plt.title(f'{ticker} Stock Price with SMA & RSI Indicators')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()

    # Show the plot
    plt.show()
```
